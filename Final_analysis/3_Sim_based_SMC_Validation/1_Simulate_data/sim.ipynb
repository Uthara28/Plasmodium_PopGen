```python
import msprime
```

```python

# Set parameters
Ne = 10 # Effective population size
sample_size = 90  # Number of samples
recom = 1e-7  # Recombination rate per base pair
mu = 0.0001 # Mutation rate per base pair
num_replicates = 1  # Number of replicates
sequence_length = 1000000  # Length of the chromosome
i = 2  # Replicate index
indx = 0  # Chromosome index (can be expanded for multiple chromosomes)

# Define demography for Kingman coalescent
demography_kingman = msprime.Demography()
demography_kingman.add_population(initial_size=Ne)

# Simulate ancestry with Kingman coalescent model
ts = msprime.sim_ancestry(
                samples=sample_size,
                sequence_length=sequence_length,
                ploidy=1,
                demography=demography_kingman,
                recombination_rate=recom)  # Add a random seed for reproducibility

print(f"Replicate {num_replicates}: Ancestry simulated for chromosome {indx+1} with length {sequence_length}")

# Simulate mutations on the ancestry
tree_sequence = msprime.sim_mutations(
                ts,
                rate=mu)  # Add a random seed for reproducibility

print(f"Replicate Mutations simulated for chromosome {indx+1}")

# Get the number of segregating sites (mutations in the sample)
num_segregating_sites = tree_sequence.num_sites
print("Number of segregating sites:", num_segregating_sites)
```
